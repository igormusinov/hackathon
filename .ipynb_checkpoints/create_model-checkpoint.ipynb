```python
data_len_for_train = raw_data.shape[0]
df = raw_data.loc[raw_data['0_ID']=='TEA'].reset_index().drop('index', axis=1)[:data_len_for_train]
```

```python
# Установим границу на Y (max value) == 3 
df = df.loc[df.Y<3]
```

```python
tmp = df.copy()[-100000:]
```

```python
'''def get_max_bid(x):
    cols_p = [c for c in x.index if 'BID_P' in c]
    cols_v = [c for c in x.index if 'BID_V' in c]
    return max(list(zip(x[cols_p].values, x[cols_v].values)), key=lambda x: x[1])[0]
def get_max_ask(x):
    cols_p = [c for c in x.index if 'ASK_P' in c]
    cols_v = [c for c in x.index if 'ASK_V' in c]
    return max(list(zip(x[cols_p].values, x[cols_v].values)), key=lambda x: x[1])[0]'''

cols_bid_p = [c for c in tmp.columns if 'BID_P' in c]
cols_bid_v = [c for c in tmp.columns if 'BID_V' in c]
cols_ask_p = [c for c in tmp.columns if 'ASK_P' in c]
cols_ask_v = [c for c in tmp.columns if 'ASK_V' in c]

tmp['bid_ask_sum'] = (tmp[cols_bid_p].values * tmp[cols_bid_v].values).sum(axis=1) / (tmp[cols_ask_p].values * tmp[cols_ask_v].values).sum(axis=1)
tmp['bid_std'] = tmp[cols_bid_p]
```

```python
(tmp[cols_bid_p].sum(axis=1)-tmp[cols_ask_p].sum(axis=1)).value_counts()##.max() - tmp[cols_bid_p].min()
```

```python
X = tmp[['BID_P_2','BID_P_3','BID_P_4','ASK_P_2','ASK_P_3','ASK_P_4','bid_ask_max']].values
Y = tmp[['Y']].values
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# fit model no training data
model = xgb.XGBRegressor(learning_rate=0.05, max_depth=6)
model.fit(X, Y)
```

```python
# make predictions for test data
y_pred = model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(ytrue, ypred):
    return np.sqrt(mean_squared_error(ytrue, ypred))
```

```python
score_test = rmse(Y, y_pred)
    
print(10/score_test)
```

```python
import pickle

model.n_jobs = model.nthread
pickle.dump(model, open('./mlanarch/my_model.txt', 'wb'))
```
